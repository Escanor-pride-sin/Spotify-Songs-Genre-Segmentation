```python
#Computing Cosine Similarity Matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(scaled_data)
```

Calculated pairwise cosine similarity between all songs in the dataset.

```python
#Building a Recommendation Function
```

```python
#Function to recommend songs based on similarity
```

```python
def recommend_songs(song_index, cosine_sim=cosine_sim, data=data, top_n=5):
     # Get similarity scores for the input song
    sim_scores = list(enumerate(cosine_sim[song_index]))
    sim_scores1 = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get the indices of the most similar song
    top_indices = [i[0] for i in sim_scores[1:top_n+1]]
    # Return the top N most similar songs
    recommended_songs = data.iloc[top_indices][['track_name', 'track_artist', 'playlist_name','track_popularity']]
    return recommended_songs
```

Built a function that takes a song index as input and recommends the top N most similar songs.
